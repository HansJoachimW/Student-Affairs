```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import re
from collections import Counter
```

```python
df = pd.read_csv('scraped_data.csv', encoding='utf-8', encoding_errors='ignore', delimiter=",")
df.head()
```

```python
df = df[['alt', 'caption', 'timestamp', 'url']]
df.head()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date

df.head()
```

```python
keywords = ['mahasiswa', 'college', 'universitas', 'university', 'kompetisi', 'competition', 'nasional', 'lomba', 'olimpiade', 'olympiad']

df = df[df['caption'].str.contains('|'.join(keywords), case=False, na=False)]

df.head()
```

```python
df.shape
```

```python
def extract_competition_name(caption):
    keywords_pattern = '|'.join([
        'mahasiswa', 'college', 'universitas', 'university',
        'kompetisi', 'competition', 'nasional', 'lomba',
        'olimpiade', 'olympiad'
    ])

    pattern = fr'(\w+(?:\s+\w+)*)\s+({keywords_pattern})\s+(\w+(\s+\w+)*)'
    
    match = re.search(pattern, caption, re.IGNORECASE)
    
    if match:
        preceding_words = match.group(1)
        keyword = match.group(2)
        competition_name = match.group(3)
        return {
            'preceding_words': preceding_words,
            'keyword': keyword,
            'competition_name': competition_name
        }
    return None
```

```python
df['competition_name'] = df['caption'].apply(extract_competition_name)

df['competition_name'].head()
```
