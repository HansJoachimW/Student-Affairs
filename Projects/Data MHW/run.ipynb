```python
import pandas as pd
```

```python
df_results = pd.read_csv('results.csv', delimiter = ';')
df_users = pd.read_csv('users.csv', delimiter = ';')
```

```python
print("First CSV:")
df_results = df_results.drop(columns=['id'])
df_results.head()
```

```python
print("\nSecond CSV:")
df_users = df_users.drop(columns=['password', 'role', 'created_at', 'updated_at'])
df_users.head()
```

```python
merged_df = pd.merge(df_results, df_users, left_on='user_id', right_on='id', how='left')
```

```python
merged_df = merged_df.drop(columns=['id', 'user_id'])
```

```python
columns = ['name', 'email'] + [col for col in merged_df.columns if col not in ['name', 'email']]
merged_df = merged_df[columns]

merged_df.head()
```

```python
merged_df = merged_df.rename(columns={'quiz_id': 'quiz'})

merged_df['quiz'] = merged_df['quiz'].replace({1: 'Pretest', 2: 'Posttest'})

merged_df.head()
```

```python
merged_df = merged_df.sort_values(by=['name', 'quiz', 'updated_at'], ascending=[True, False, False])
```

```python
final_df = merged_df.drop_duplicates(subset=['name', 'quiz'], keep='first')

final_df.head()
```

```python
print("Final DF:")
final_df = final_df.drop(columns=['created_at', 'updated_at'])
final_df.head()
```

```python
results_0_to_13 = final_df[(final_df['result'] >= 0) & (final_df['result'] <= 13)]
results_above_13 = final_df[final_df['result'] > 13]
```

```python
output_file = 'mhw_quiz_results.xlsx'
with pd.ExcelWriter(output_file) as writer:
    final_df.to_excel(writer, sheet_name='All', index=False)
    results_0_to_13.to_excel(writer, sheet_name='Results_0_to_13', index=False)
    results_above_13.to_excel(writer, sheet_name='Results_above_13', index=False)

print(f"Final DataFrame saved to {output_file} with two sheets.")
```
