```python
import pandas as pd
```

```python
student_pretest = pd.read_excel('PFA Mahasiswa - Survey 1 (Responses).xlsx', sheet_name="Form Responses 1")
student_posttest = pd.read_excel('PFA Mahasiswa - Survey 2 (Responses).xlsx', sheet_name="Form Responses 1")
lecturer_pretest = pd.read_excel('PFA Dosen - Survey 1 (Responses).xlsx', sheet_name="Form Responses 1")
lecturer_posttest = pd.read_excel('PFA Dosen - Survey 2 (Responses).xlsx', sheet_name="Form Responses 1")
```

```python
student_nim_pretest = set(student_pretest['NIM'])
student_nim_posttest = set(student_posttest['NIM'])
student_completers = student_nim_pretest.intersection(student_nim_posttest)
```

```python
lecturer_nik_pretest = set(lecturer_pretest['NIK'])
lecturer_nik_posttest = set(lecturer_posttest['NIK'])
lecturer_completers = lecturer_nik_pretest.intersection(lecturer_nik_posttest)
```

```python
student_pretest = student_pretest[student_pretest['NIM'].isin(student_completers)]
student_posttest = student_posttest[student_posttest['NIM'].isin(student_completers)]
lecturer_pretest = lecturer_pretest[lecturer_pretest['NIK'].isin(lecturer_completers)]
lecturer_posttest = lecturer_posttest[lecturer_posttest['NIK'].isin(lecturer_completers)]
```

```python
student_pretest['Total_Score'] = student_pretest.iloc[:, 5:].sum(axis=1)
student_posttest['Total_Score'] = student_posttest.iloc[:, 5:15].sum(axis=1)
lecturer_pretest['Total_Score'] = lecturer_pretest.iloc[:, 4:].sum(axis=1)
lecturer_posttest['Total_Score'] = lecturer_posttest.iloc[:, 4:14].sum(axis=1)
```

```python
student_pretest['Test_Type'] = 'Pretest'
student_posttest['Test_Type'] = 'Posttest'
lecturer_pretest['Test_Type'] = 'Pretest'
lecturer_posttest['Test_Type'] = 'Posttest'
```

```python
student_summary = pd.merge(
    student_pretest, 
    student_posttest, 
    on=['NIM', 'Nama Lengkap'], 
    suffixes=('_Pretest', '_Posttest')
)
```

```python
lecturer_summary = pd.merge(
    lecturer_pretest, 
    lecturer_posttest, 
    on=['NIK', 'Nama Lengkap (beserta gelar)'], 
    suffixes=('_Pretest', '_Posttest')
)
```
