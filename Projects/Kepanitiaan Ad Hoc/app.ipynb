```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load your Excel file
df = pd.read_excel('All Identifikasi KP activity panitia ad hoc.xlsx', sheet_name='Sheet0')
df.head()
```

```python
keywords = ['Koordinator', 'Ketua', 'Wakil Ketua', 'Sekretaris', 'Bendahara', 'Panitia Ad Hoc']

def update_status(row):
    if pd.notna(row['description']):
        for keyword in keywords:
            if keyword in row['description']:
                return keyword
    return row['status']

df['status'] = df.apply(update_status, axis=1)

unique_activities = df.drop_duplicates(subset='activity')[['code', 'activity', 'trx_code']]

no_description = df[df['description'].isna()]

with pd.ExcelWriter('output_file.xlsx', engine='xlsxwriter') as writer:
    unique_activities.to_excel(writer, sheet_name='Unique Activities', index=False)

    no_description.to_excel(writer, sheet_name='No Description', index=False)
    
    for keyword in keywords:
        filtered_df = df[df['status'] == keyword]
        if not filtered_df.empty:
            filtered_df.to_excel(writer, sheet_name=keyword, index=False)
    
    df.to_excel(writer, sheet_name='All', index=False)
```
