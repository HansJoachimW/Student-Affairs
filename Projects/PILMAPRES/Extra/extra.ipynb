```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import re
from collections import Counter
```

```python
df = pd.read_csv('igdata.csv', encoding='utf-8', encoding_errors='ignore', delimiter=",")
df.head()
```

```python
universities_df = pd.read_csv('UniList.csv', encoding='utf-8', encoding_errors='ignore')
universities_list = universities_df['Name'].str.strip().str.lower().tolist()
```

```python
universities_list
```

```python
def extract_and_clean_universities(text):
    pattern = r'\bUniversitas\s[A-Z][a-zA-Z\s]*|University\s[A-Z][a-zA-Z\s]*\b'
    universities = re.findall(pattern, text)
    # Clean the names by stripping whitespace, newlines, and normalizing spaces
    universities = [uni.strip().replace('\n', ' ').replace('  ', ' ') for uni in universities]
    return universities
```

```python
all_text = df['alt'].fillna('') + " " + df['caption'].fillna('')
```

```python
all_text
```

```python
df['universities'] = all_text.apply(extract_and_clean_universities)
exploded_data = df.explode('universities').dropna(subset=['universities'])
exploded_data['universities'] = exploded_data['universities'].str.lower().str.strip()
```

```python
exploded_data
```

```python
exploded_data['universities'] = exploded_data['universities'].apply(lambda x: x if x in universities_list else None)
```

```python
exploded_data['universities']
```

```python
filtered_data = exploded_data.dropna(subset=['universities'])
```

```python
university_counts = filtered_data.groupby('universities').size().reset_index(name='count')
university_counts = university_counts.sort_values(by='count', ascending=False)
```

```python
university_counts['universities'] = university_counts['universities'].str.title()
```

```python
university_counts
```

```python
total_count = university_counts['count'].sum()
print(f"Total count of university mentions: {total_count}")
```

```python
word_count = all_text.str.lower().str.count(r'\buniversitas\b').sum()

print(f"Total occurrences of 'universitas': {word_count}")
```
