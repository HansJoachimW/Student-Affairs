```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import calendar
```

Kompetisi


```python
df_kompetisi = pd.read_csv('csv/data_kompetisi.csv', encoding='utf-8', encoding_errors='ignore', delimiter=";", dtype={'NIS': str})
df_kompetisi.head()
```

```python
df_kompetisi = df_kompetisi[df_kompetisi['Score'] != 0]
```

```python
df_kompetisi['Start Date'] = pd.to_datetime(df_kompetisi['Start Date'], errors='coerce', format='%d/%m/%Y')
df_kompetisi['End Date'] = pd.to_datetime(df_kompetisi['End Date'], errors='coerce', format='%d/%m/%Y')

df_kompetisi = df_kompetisi.dropna(subset=['Start Date', 'End Date'])
```

```python
df_kompetisi.head()
```

```python
start_range = '2023-01-01'
end_range = '2023-07-31'

# Filter the DataFrame where Start Date and End Date fall between January and July
df_kompetisi = df_kompetisi[(df_kompetisi['End Date'] >= start_range) & (df_kompetisi['End Date'] <= end_range)]
```

```python
# levels = 'External Regional|External International'
# df = df[~df['Level'].str.contains(levels, na=False)]
```

```python
df_kompetisi['Score'] = pd.to_numeric(df_kompetisi['Score'], errors='coerce')
```

```python
df_kompetisi = df_kompetisi.drop(columns=['NIS', 'Student Year', 'Period'])
```

```python
df_kompetisi = df_kompetisi.dropna()
```

```python
df_kompetisi_teams = df_kompetisi[df_kompetisi['Team Category'] == 'Team'].drop_duplicates(subset=['Activity', 'Major', 'Criteria','Start Date', 'End Date'])
df_kompetisi_individuals = df_kompetisi[df_kompetisi['Team Category'] == 'Individual']
```

```python
df_kompetisi_teams
```

```python
df_kompetisi_combined = pd.concat([df_kompetisi_teams, df_kompetisi_individuals])
df_kompetisi_combined
```

```python
df_kompetisi_combined['Year'] = df_kompetisi_combined['End Date'].dt.year
df_kompetisi_combined['Month'] = df_kompetisi_combined['End Date'].dt.month.apply(lambda x: calendar.month_name[x])
```

```python
df_kompetisi_scores = df_kompetisi_combined.groupby(['Major', 'Month', 'Year', 'Team Category', 'Status', 'Level'])['Score'].sum().reset_index()
df_kompetisi_scores
```

```python
output_file_path_kompetisi = 'C:\\Users\\hansj\\OneDrive\\Documents\\Student Affairs\\Projects\\SIMKATMAWA\\output\\output_kompetisi.xlsx'
df_kompetisi_scores.to_excel(output_file_path_kompetisi, index=False)
```

Nonlomba


```python
df_nonlomba = pd.read_csv('csv/data_nonlomba.csv', encoding='utf-8', encoding_errors='ignore', delimiter=";", dtype={'NIS': str})
df_nonlomba.head()
```

```python
df_nonlomba = df_nonlomba[df_nonlomba['Score'] != 0]
```

```python
df_nonlomba.drop(columns=['Website', 'Certificate', 'Assignment Letter', 'Report', 'Documentation', 'Organizer'], inplace=True)
```

```python
df_nonlomba = df_nonlomba[df_nonlomba['Status'] != 'Narasumber / Pemateri Acara Seminar / Workshop / Pemakalah']
```

```python
df_nonlomba['Start Date'] = pd.to_datetime(df_nonlomba['Start Date'], errors='coerce')
df_nonlomba['End Date'] = pd.to_datetime(df_nonlomba['End Date'], errors='coerce')

df_nonlomba = df_nonlomba.dropna(subset=['Start Date', 'End Date'])
```

```python
start_range = '2023-01-01'
end_range = '2023-07-31'

# Filter the DataFrame where Start Date and End Date fall between January and July
df_nonlomba = df_nonlomba[(df_nonlomba['End Date'] >= start_range) & (df_nonlomba['End Date'] <= end_range)]
```

```python
df_nonlomba['Score'] = pd.to_numeric(df_nonlomba['Score'], errors='coerce')
```

```python
df_nonlomba['Year'] = df_nonlomba['End Date'].dt.year
df_nonlomba['Month'] = df_nonlomba['End Date'].dt.month.apply(lambda x: calendar.month_name[x])
```

```python
df_nonlomba_scores = df_nonlomba.groupby(['Major', 'Month', 'Year', 'Status', 'Level'])['Score'].sum().reset_index()
df_nonlomba_scores
```

```python
output_file_path_nonlomba = 'C:\\Users\\hansj\\OneDrive\\Documents\\Student Affairs\\Projects\\SIMKATMAWA\\output\\output_nonlomba.xlsx'
df_nonlomba_scores.to_excel(output_file_path_nonlomba, index=False)
```
