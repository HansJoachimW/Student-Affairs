```python
import os
from PyPDF2 import PdfReader, PdfWriter

def split_pdf(input_pdf, output_folder="certificates/"):
    """
    Splits a PDF into individual pages and saves them in the specified folder.

    Args:
        input_pdf (str): Path to the input PDF file.
        output_folder (str): Path to the folder where individual pages will be saved.
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    reader = PdfReader(input_pdf)
    for page_num, page in enumerate(reader.pages):
        writer = PdfWriter()
        writer.add_page(page)
        output_path = os.path.join(output_folder, f"page_{page_num + 1}.pdf")
        with open(output_path, "wb") as output_file:
            writer.write(output_file)
        print(f"Page {page_num + 1} saved to {output_path}")

# Example usage:
input_pdf_path = "Sertifikat Pemenang Lomba.pdf"  # Replace with your PDF file path
output_folder_path = "split_pages"  # Replace with your desired output folder path

split_pdf(input_pdf_path, output_folder_path)
```
