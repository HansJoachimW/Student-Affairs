```python
import os
from PyPDF2 import PdfMerger
```

```python
def combine_pdfs(pdf_list, output_file):
    """
    Combine multiple PDF files into a single PDF.

    :param pdf_list: List of paths to the PDF files to be combined.
    :param output_file: Path to the output combined PDF file.
    """
    merger = PdfMerger()

    for pdf in pdf_list:
        try:
            merger.append(pdf)
            print(f"Added: {pdf}")
        except Exception as e:
            print(f"Error adding {pdf}: {e}")

    try:
        merger.write(output_file)
        merger.close()
        print(f"Combined PDF saved as: {output_file}")
    except Exception as e:
        print(f"Error writing combined PDF: {e}")
```

```python
input_dir = "input/"  # Replace with your directory path
output_dir = "output/"  # Replace with your desired output file name
```

```python
output_file = os.path.join(output_dir, "icoen.pdf")
```

```python
os.makedirs(output_dir, exist_ok=True)

# Collect all PDF files from the directory
pdf_list = [os.path.join(input_dir, f) for f in os.listdir(input_dir) if f.endswith('.pdf')]
pdf_list.sort()  # Optional: Sort PDFs alphabetically
```

```python
if pdf_list:
    combine_pdfs(pdf_list, output_file)
else:
    print("No PDF files found in the specified directory.")
```
