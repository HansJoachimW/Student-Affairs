```python
df = df.drop(columns=["period", "nama_survey", "question_name", "tanggal_input"])
```

```python
dass_mapping = {
    "Depression": ["Q3.3", "Q3.5", "Q3.10", "Q3.13"],
    "Anxiety": ["Q3.2", "Q3.4", "Q3.7", "Q3.9"],
    "Stress": ["Q3.1", "Q3.6", "Q3.8", "Q3.11", "Q3.12", "Q3.14"]
}

abusi_mapping = {
    "Pikiran untuk NSSI": ["Q4.4", "Q4.5", "Q4.6", "Q4.7"],
    "Perilaku NSSI": ["Q4.8", "Q4.9", "Q4.10"]
}
```

```python
relevant_questions = ["Q3.1", "Q3.2", "Q3.3", "Q3.4", "Q3.5", "Q3.6", "Q3.7", "Q3.8", "Q3.9", "Q3.10", "Q3.11", "Q3.12", "Q3.13", "Q3.14", "Q4.1","Q4.2","Q4.3","Q4.4", "Q4.5", "Q4.6", "Q4.7", "Q4.8", "Q4.9", "Q4.10", "Q4.11", "Q4.12", "Q4.13", "Q4.14", "Q4.15"]
df = df[df["question_id"].isin(relevant_questions)]
```

```python
# df["answer"] = df["answer"].astype(str)
```

```python
mapped_questions = set(sum(dass_mapping.values(), []) + sum(abusi_mapping.values(), []))

def replace_answer(row):
    if (
        row["answer"] == "0"
        and row["question_id"] not in mapped_questions
        and pd.notna(row["answer_text"])
        and str(row["answer_text"]).strip() != ""
    ):
        return str(row["answer_text"])  # Kembalikan jawaban yang benar
    return row["answer"]  # Biarkan tetap sama jika tidak memenuhi kondisi

df["answer"] = df.apply(replace_answer, axis=1)

```

```python
df.head(29)
```

```python
def classify_dass(value, category):
    if category == "Depression":
        if value < 3.20: return "Normal"
        elif 3.20 <= value < 6.40: return "Mild"
        elif 6.40 <= value < 9.60: return "Moderate"
        elif 9.60 <= value < 12.80: return "Severe"
        else: return "Extremely Severe"
        
    elif category == "Anxiety":
        if value < 4.80: return "Normal"
        elif 4.80 <= value < 9.60: return "Mild"
        elif 9.60 <= value < 14.40: return "Moderate"
        elif 14.40 <= value < 19.20: return "Severe"
        else: return "Extremely Severe"
    
    elif category == "Stress":
        if value < 4.80: return "Normal"
        elif 4.80 <= value < 9.60: return "Mild"
        elif 9.60 <= value < 14.40: return "Moderate"
        elif 14.40 <= value < 19.20: return "Severe"
        else: return "Extremely Severe"

def classify_abusi(value, category):
    if category == "Pikiran untuk NSSI":
        if value < 5.33: return "Normal"
        elif 5.33 <= value < 10.66: return "Moderate"
        else: return "Severe"
    
    elif category == "Perilaku NSSI":
        if value < 6.67: return "Normal"
        elif 6.67 <= value < 13.34: return "Moderate"
        else: return "Severe"
```

```python
def compute_sums(df, mapping):
    scores = {}
    
    for category, questions in mapping.items():
        filtered_df = df[df["question_id"].isin(questions)].copy()  # Salin data yang difilter
        filtered_df["answer"] = pd.to_numeric(filtered_df["answer"], errors="coerce").fillna(0).astype(int)  # Konversi ke int
        scores[category] = filtered_df.groupby("nis")["answer"].sum()
    
    return pd.DataFrame(scores).reset_index()
```

```python
dass_df = compute_sums(df, dass_mapping)
abusi_df = compute_sums(df, abusi_mapping)
```

```python
dass_df.iloc[:, 1:] = dass_df.iloc[:, 1:].apply(pd.to_numeric, errors="coerce").fillna(0)
abusi_df.iloc[:, 1:] = abusi_df.iloc[:, 1:].apply(pd.to_numeric, errors="coerce").fillna(0)
```

```python
for col in dass_df.columns[1:]:  # Skip 'nis'
    dass_df[f"{col} Classification"] = dass_df[col].apply(lambda x: classify_dass(x, col))

for col in abusi_df.columns[1:]:  # Skip 'nis'
    abusi_df[f"{col} Classification"] = abusi_df[col].apply(lambda x: classify_abusi(x, col))
```

```python
# Get unique student information
student_info = df[["nis", "name", "prodi", "faculty"]].drop_duplicates(subset="nis")

# Merge with DASS and Abusi data
result = student_info.merge(dass_df, on="nis", how="left").merge(abusi_df, on="nis", how="left")
```

```python
q3_q4_answers = df[df["question_id"].str.startswith("Q3.") | df["question_id"].str.startswith("Q4.")][["nis", "question_id", "answer"]]
```

```python
checklist_questions = ["Q4.1", "Q4.3", "Q4.15"]
q3_q4_answers = q3_q4_answers.groupby(["nis", "question_id"]).agg(lambda x: ", ".join(map(str, x))).reset_index()
```

```python
q3_q4_pivot = q3_q4_answers.pivot(index="nis", columns="question_id", values="answer").reset_index()
```

```python
final_result = result.merge(q3_q4_pivot, on="nis", how="left")
```

```python
dass_classifications = [f"{category} Classification" for category in dass_mapping.keys()]
abusi_classifications = [f"{category} Classification" for category in abusi_mapping.keys()]
```

```python
question_columns = sorted([col for col in final_result.columns if col.startswith("Q3.") or col.startswith("Q4.")], key=lambda x: (x[:2], int(x[3:])))
column_order = (
    ["nis", "name", "prodi", "faculty"] + 
    question_columns + 
    list(dass_mapping.keys()) + dass_classifications + 
    list(abusi_mapping.keys()) + abusi_classifications
)
```
